## 🔠 Métodos upper() y lower()

### 📌 Explicación del código

`titanic["name"].str.upper()`

🔹 **`titanic["name"]`**
Selecciona la columna **`name`** del DataFrame.

🔹 **`.str`**
Permite aplicar funciones de texto (strings) a cada valor de la columna.

🔹 **`.upper()`**
Convierte todas las cadenas a **mayúsculas**.

📌 ¿Qué hace en conjunto?

Transforma cada nombre del DataFrame a su versión en mayúsculas.

🔹 **Resultado**
Una **Serie** con los nombres en formato MAYÚSCULAS.

👉 **En resumen:**
Convierte todos los nombres de los pasajeros a letras mayúsculas.

```python
titanic["name"].str.upper()
```

```python
titanic["name"]
```

### 📌 Explicación del código

`titanic["lower_name"] = titanic["name"].str.lower()`

🔹 **`titanic["name"]`**
Selecciona la columna **`name`**.

🔹 **`.str.lower()`**
Convierte cada texto a **minúsculas**.

🔹 **`titanic["lower_name"] = ...`**
Crea una **nueva columna** llamada `lower_name` y guarda ahí el resultado.

📌 ¿Qué hace en conjunto?

- Toma los nombres originales
- Los convierte a minúsculas
- Los guarda en una nueva columna

🔹 **Resultado**
El DataFrame ahora tiene una columna adicional con los nombres en minúsculas.

👉 **En resumen:**
Crea una nueva columna con los nombres convertidos a letras minúsculas.

```python
titanic["lower_name"] = titanic["name"].str.lower()
```

```python
titanic
```

### 📌 Explicación del código

`titanic.lower_name.str.capitalize()`

🔹 **`titanic.lower_name`**
Accede a la columna **`lower_name`** del DataFrame (nombres en minúsculas).

🔹 **`.str.capitalize()`**
Convierte cada cadena para que:
- La **primera letra sea mayúscula**
- El resto quede en **minúsculas**

📌 ¿Qué hace en conjunto?

Toma los nombres en minúsculas y los transforma a formato tipo:
`Juan`, `Maria`, etc.

🔹 **Resultado**
Devuelve una **Serie** con los nombres capitalizados (no modifica el DataFrame a menos que se reasigne).

👉 **En resumen:**
Convierte cada nombre a formato con **inicial mayúscula y resto en minúsculas**.

```python
titanic.lower_name.str.capitalize()
```

## String Indexing

### 📌 Explicación del código

`titanic["cabin"].str[0]`

🔹 **`titanic["cabin"]`**
Selecciona la columna **`cabin`** (cabina de cada pasajero).

🔹 **`.str`**
Permite aplicar operaciones de texto a cada valor.

🔹 **`[0]`**
Obtiene el **primer carácter** de cada cadena.

📌 ¿Qué hace en conjunto?

Extrae la **primera letra de la cabina**, que generalmente indica la **sección o cubierta del barco** (por ejemplo: A, B, C, etc.).

🔹 **Resultado**
Devuelve una **Serie** con la primera letra de cada cabina.

⚠️ Si hay valores nulos (`NaN`), el resultado también será `NaN`.

👉 **En resumen:**
Obtiene la **letra inicial de cada cabina** del Titanic.

```python
titanic["cabin"].str[0]
```

### 📌 Explicación del código

`titanic["deck"] = titanic["cabin"].str[0]`

🔹 **`titanic["cabin"]`**
Selecciona la columna de cabinas.

🔹 **`.str[0]`**
Extrae el **primer carácter** de cada valor (la letra de la cabina).

🔹 **`titanic["deck"] = ...`**
Crea una nueva columna llamada **`deck`** y guarda ahí ese valor.

📌 ¿Qué hace en conjunto?

- Toma la cabina de cada pasajero
- Extrae la primera letra (por ejemplo: A, B, C…)
- La guarda en una nueva columna

🔹 **Resultado**
El DataFrame ahora incluye una columna `deck` que indica la **cubierta del barco**.

⚠️ Si hay valores nulos en `cabin`, también aparecerán como `NaN`.

👉 **En resumen:**
Crea una nueva columna con la **letra inicial de la cabina**, representando la cubierta.

```python
titanic["deck"] = titanic["cabin"].str[0]
```

```python
titanic
```

### 📊 Explicación del código

`titanic.groupby("deck").mean(numeric_only=True)`

🔹 **`groupby("deck")`**
Agrupa los datos del DataFrame según la columna **`deck`** (la cubierta del barco: A, B, C, etc.).

🔹 **`.mean(numeric_only=True)`**
Calcula el **promedio de todas las columnas numéricas** dentro de cada grupo.

📌 ¿Qué hace en conjunto?

- Agrupa a los pasajeros según la **cubierta donde estaban**
- Calcula promedios como:
  - edad promedio
  - tarifa promedio
  - tasa de supervivencia promedio

🔹 **Resultado**
Un DataFrame donde:
- Índice → `deck`
- Valores → promedios de las variables numéricas por cubierta

👉 **En resumen:**
Calcula estadísticas promedio de los pasajeros agrupándolos por **cubierta del Titanic**.

```python
titanic.groupby("deck").mean(numeric_only=True)
```

### 📊 Explicación del código

`titanic.groupby("deck")["survived"].mean().sort_values().plot(kind="bar")`

🔹 **`groupby("deck")`**
Agrupa los datos por la columna **`deck`** (cubierta del barco).

🔹 **`["survived"]`**
Selecciona la columna de supervivencia:
- `0` → no sobrevivió
- `1` → sobrevivió

🔹 **`.mean()`**
Calcula el promedio de `survived`, lo que equivale al **porcentaje de supervivencia** por cubierta.

🔹 **`.sort_values()`**
Ordena los resultados de menor a mayor (cubiertas con menor supervivencia primero).

🔹 **`.plot(kind="bar")`**
Genera una **gráfica de barras**.

📌 ¿Qué muestra la gráfica?

- Eje X → cubiertas (`deck`)
- Eje Y → tasa de supervivencia
- Ordenadas de menor a mayor supervivencia

👉 **En resumen:**
Calcula el porcentaje de supervivencia por cubierta, lo ordena y lo muestra en una gráfica de barras.

```python
titanic.groupby("deck")["survived"].mean().sort_values().plot(kind="bar")
```

## ✂️ Eliminar espacios con strip()

### 📌 Explicación del código

`s = pd.Series(['1. Hawk.  ', '2. Pickle!\n', '3. Melonhead?\t'])`

🔹 **`pd.Series(...)`**
Crea una **Serie de Pandas**, que es como una lista con índice.

🔹 **Contenido de la Serie**
Se almacenan 3 cadenas de texto:

- `'1. Hawk.  '` → tiene **espacios al final**
- `'2. Pickle!\n'` → contiene un **salto de línea (`\n`)**
- `'3. Melonhead?\t'` → contiene un **tabulador (`\t`)**

🔹 **Caracteres especiales**
- `\n` → nueva línea
- `\t` → tabulación
- espacios → también cuentan como parte del texto

🔹 **Resultado**
Una Serie `s` con textos que incluyen **espacios y caracteres especiales**.


```python
s = pd.Series(['1. Hawk.  ', '2. Pickle!\n', '3. Melonhead?\t'])
```

```python
s
```

### 📌 Explicación del código

`s.str.strip()`

🔹 **`s`**
Es una Serie de textos que contiene espacios y caracteres especiales.

🔹 **`.str.strip()`**
Elimina los **espacios en blanco al inicio y al final** de cada cadena.

📌 También elimina caracteres como:
- espacios `" "`
- saltos de línea `\n`
- tabulaciones `\t`

🔹 **¿Qué hace en conjunto?**

Limpia cada texto quitando caracteres innecesarios al principio y al final.

🔹 **Resultado**
Devuelve una nueva Serie con los textos más limpios, por ejemplo:

- `'1. Hawk.'`
- `'2. Pickle!'`
- `'3. Melonhead?'`

👉 **En resumen:**
Elimina espacios y caracteres especiales al inicio y final de cada texto.

```python
s.str.strip()
```

### 📌 Explicación del código

`s.str.strip(to_strip="123. \n \t")`

🔹 **`.str.strip()`**
Elimina caracteres al **inicio y al final** de cada cadena.

🔹 **`to_strip="123. \n \t"`**
Indica exactamente qué caracteres se deben eliminar:
- números: `1`, `2`, `3`
- punto: `.`
- espacio: `" "`
- salto de línea: `\n`
- tabulación: `\t`

📌 ¿Qué hace en conjunto?

Quita todos esos caracteres **solo si están al inicio o al final** del texto (no en medio).

🔹 **Ejemplo de resultado**

- `'1. Hawk.  '` → `'Hawk'`
- `'2. Pickle!\n'` → `'Pickle!'`
- `'3. Melonhead?\t'` → `'Melonhead?'`

👉 **En resumen:**
Elimina caracteres específicos (números, puntos, espacios y saltos) del inicio y final de cada texto.

```python
s.str.strip(to_strip="123. \n \t")
```

### 📌 Explicación del código

`s.str.rstrip()`

🔹 **`.str.rstrip()`**
Elimina los **espacios en blanco y caracteres especiales solo al final** de cada cadena.

📌 A diferencia de `strip()`:
- `strip()` → limpia inicio y final
- `rstrip()` → limpia solo el **lado derecho (final)**

🔹 **¿Qué elimina?**
- espacios `" "`
- saltos de línea `\n`
- tabulaciones `\t`

🔹 **Ejemplo de resultado**

- `'1. Hawk.  '` → `'1. Hawk.'`
- `'2. Pickle!\n'` → `'2. Pickle!'`
- `'3. Melonhead?\t'` → `'3. Melonhead?'`

👉 **En resumen:**
Elimina espacios y caracteres especiales **solo al final** de cada texto.

```python
s.str.rstrip()
```

### 📌 Explicación del código

`s.str.lstrip()`

🔹 **`.str.lstrip()`**
Elimina los **espacios en blanco y caracteres especiales solo al inicio** de cada cadena.

📌 A diferencia de otros métodos:
- `strip()` → inicio y final
- `rstrip()` → solo final
- `lstrip()` → solo **inicio**

🔹 **¿Qué elimina?**
- espacios `" "`
- saltos de línea `\n`
- tabulaciones `\t`

🔹 **Ejemplo de resultado**

En este caso, como las cadenas no tienen espacios al inicio, el resultado será prácticamente igual:

- `'1. Hawk.  '` → `'1. Hawk.  '`
- `'2. Pickle!\n'` → `'2. Pickle!\n'`
- `'3. Melonhead?\t'` → `'3. Melonhead?\t'`

👉 **En resumen:**
Elimina espacios y caracteres especiales **solo al inicio**, aunque en este ejemplo no hay cambios visibles.

```python
s.str.lstrip()
```

### 📌 Explicación del código

`s.str.strip(to_strip="123.")`

🔹 **`.str.strip()`**
Elimina caracteres al **inicio y al final** de cada cadena.

🔹 **`to_strip="123."`**
Indica que se deben eliminar únicamente:
- `1`, `2`, `3`
- `.` (punto)

📌 ¿Qué hace en conjunto?

Quita esos caracteres si aparecen **al inicio o al final**, pero **no elimina espacios ni `\n` o `\t`** porque no están incluidos.

🔹 **Ejemplo de resultado**

- `'1. Hawk.  '` → `' Hawk.  '`
- `'2. Pickle!\n'` → `' Pickle!\n'`
- `'3. Melonhead?\t'` → `' Melonhead?\t'`

👉 **En resumen:**
Elimina solo números y puntos al inicio y final, dejando intactos los espacios y caracteres especiales.

```python
s.str.strip(to_strip="123.")
```
