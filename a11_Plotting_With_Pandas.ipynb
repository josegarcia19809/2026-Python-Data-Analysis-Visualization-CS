### 📌 Explicación del código

`plt.style.use("fivethirtyeight")`

🔹 **`plt`**
Hace referencia a la librería **Matplotlib (pyplot)**, usada para crear gráficas.

🔹 **`style.use()`**
Permite aplicar un **estilo visual predefinido** a las gráficas.

🔹 **`"fivethirtyeight"`**
Es un estilo inspirado en el sitio de análisis de datos **FiveThirtyEight**, conocido por gráficas limpias y modernas.

📌 ¿Qué cambia?

- Colores más suaves
- Fondos grises
- Líneas más visibles
- Mejor presentación general

🔹 **Resultado**
Todas las gráficas que hagas después tendrán ese estilo automáticamente.

👉 **En resumen:**
Aplica un estilo visual moderno a las gráficas de Matplotlib.

```python
plt.style.use("fivethirtyeight")
```

### 📊 Explicación del código

`titanic.sex.value_counts().plot(kind="bar")`

🔹 **`titanic.sex`**
Accede a la columna **`sex`** del DataFrame.

🔹 **`.value_counts()`**
Cuenta cuántas veces aparece cada valor (por ejemplo: `male`, `female`).

🔹 **`.plot(kind="bar")`**
Crea una **gráfica de barras** con esos conteos.

📌 ¿Qué muestra la gráfica?

- Eje X → categorías (`male`, `female`)
- Eje Y → cantidad de pasajeros
- Cada barra representa cuántos hay de cada sexo

👉 **En resumen:**
Cuenta cuántos hombres y mujeres hay en el Titanic y lo muestra en una gráfica de barras.

```python
titanic.sex.value_counts().plot(kind="bar")
```

```python
ufos["month"] = ufos["date_time"].dt.month
ufos["year"] = ufos["date_time"].dt.year
```

```python
sightings = ufos.month.value_counts().sort_index()
sightings.plot(
    kind="bar", 
    title="UFO Sightings By Month", 
    xlabel="Month",
    ylabel="Num Sightings",
    fontsize="10")
```

```python
sightings = ufos.month.value_counts().sort_index()
sightings.plot(
    kind="bar", 
    title="UFO Sightings By Month", 
    fontsize="10")
plt.xlabel("Month", fontsize=24, color="olive")
plt.ylabel("Num Sightings", fontsize=24)
```

```python
sightings = ufos.month.value_counts().sort_index()
sightings.plot(
    kind="bar", 
    title="UFO Sightings By Month", 
    fontsize="10")
plt.xlabel("Month", fontsize=24, color="olive")
plt.ylabel("Num Sightings", fontsize=24)
```

## Manual Subplots

```python
fig, axs = plt.subplots(1,2)
ufos.year.value_counts().sort_index().plot(ax=axs[0])
axs[0].set_title("Sightings By Year")
ufos.month.value_counts().sort_index().plot(kind="bar", ax=axs[1])
axs[1].set_title("Sightings By Month")
```

```python
months = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 
          7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
fig, axs = plt.subplots(2, 3, figsize=(14, 8))
fig.suptitle("UFO sightings by month", fontsize=14)
fig.tight_layout(pad=2)
ufos[ufos.year == 2014].month.value_counts().sort_index().rename(months).plot(kind="bar", ax=axs[0][0], title="2014")
ufos[ufos.year == 2015].month.value_counts().sort_index().rename(months).plot(kind="bar", ax=axs[0][1], title="2015")
ufos[ufos.year == 2016].month.value_counts().sort_index().rename(months).plot(kind="bar", ax=axs[0][2], title="2016")
ufos[ufos.year == 2017].month.value_counts().sort_index().rename(months).plot(kind="bar", ax=axs[1][0], title="2017")
ufos[ufos.year == 2018].month.value_counts().sort_index().rename(months).plot(kind="bar", ax=axs[1][1], title="2018")
ufos[ufos.year == 2019].month.value_counts().sort_index().rename(months).plot(kind="bar", ax=axs[1][2], title="2019")
plt.tight_layout()
```
