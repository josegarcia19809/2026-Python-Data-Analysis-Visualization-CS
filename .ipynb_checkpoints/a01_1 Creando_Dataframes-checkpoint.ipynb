# 📊 Ejemplos de creación de Dataframes usando Pandas

Este notebook muestra cómo crear diferentes DataFrames usando listas en Python.

```python
import pandas as pd
```

## 👩‍🎓 Calificaciones de estudiantes

Creamos 4 listas por separado: una para los nombres de los estudiantes, dos para sus calificaciones en Matemáticas e Historia, y otra para indicar si aprobaron. Luego combinamos estas listas en un DataFrame utilizando `pd.DataFrame`.

```python
# Listas separadas
nombres = ['Ana', 'Luis', 'Pedro']
matematicas = [85, 90, 78]
historia = [88, 76, 91]
aprobado = [True, True, True]

# Crear el DataFrame con las listas
df_calificaciones = pd.DataFrame({
    'Nombre': nombres,
    'Matemáticas': matematicas,
    'Historia': historia,
    'Aprobado': aprobado
})

df_calificaciones
```

## 💰 Ventas Diarias

En este ejemplo, usaremos **listas separadas** para representar los datos de ventas de distintos productos en una tienda. Cada lista representa una columna del DataFrame:

- `productos`: contiene los nombres de los productos vendidos.
- `precios`: indica el precio unitario de cada producto.
- `unidades`: muestra cuántas unidades se vendieron por producto.
- `fechas`: almacena la fecha en que se realizó cada venta.

Después combinamos todas estas listas para construir un DataFrame con `pandas`.

| Producto | Precio   | Unidades | Fecha      |
|----------|----------|----------|------------|
| Laptop   | 15000.50 | 3        | 2025-06-28 |
| Mouse    | 350.75   | 10       | 2025-06-28 |
| Teclado  | 700.00   | 5        | 2025-06-28 |



## 📋 Datos de Empleados

En este ejemplo creamos un DataFrame de Pandas a partir de una lista de listas. Cada sublista representa un empleado con sus datos: ID, nombre, departamento y años de experiencia.

Primero definimos la lista `datos_empleados` que contiene los registros de cada empleado. Luego definimos una lista `columnas_empleados` que especifica los nombres de las columnas para el DataFrame. Finalmente, usamos `pd.DataFrame` para crear el DataFrame usando esos datos y columnas.

| ID      | Nombre      | Departamento   | Años_Experiencia |
|---------|-------------|----------------|------------------|
| EMP001  | María López | Contabilidad   | 2                |
| EMP002  | Juan Pérez  | Sistemas       | 5                |
| EMP003  | Luisa Díaz  | Marketing      | 3                |


```python
datos_empleados = [
    ['EMP001', 'María López', 'Contabilidad', 2],
    ['EMP002', 'Juan Pérez', 'Sistemas', 5],
    ['EMP003', 'Luisa Díaz', 'Marketing', 3]
]
columnas_empleados = ['ID', 'Nombre', 'Departamento', 'Años_Experiencia']
df_empleados = pd.DataFrame(datos_empleados, columns=columnas_empleados)
df_empleados
```
